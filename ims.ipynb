```python
# Add New Products and
# Update details of Existing Products

import json

fd = open("records.json", "r")
txt = fd.read()
fd.close()

my_record = json.loads(txt)

choice = int(input("Enter 1 to Add new product OR Enter 2 to Update details of existing product \nEnter your choice: "))

if choice == 1:
    product_id = int(list(my_record.keys())[-1]) +1
    print("\nThe Product ID will be:", product_id)
    print("Enter the Remaining Details...")
    product = input("Enter Product Name:       ")
    size =    input("Enter Product Size:       ")
    gender =  input("Enter Product for Gender: ")
    price =   input("Enter Product Price:      ")
    stock =   input("Enter Product Stock:      ")

    my_record[product_id] = {
        "product": product,
        "size": size,
        "gender": gender,
        "price": price,
        "stock": stock
    }

    js = json.dumps(my_record)

    fd = open("records.json", "w")
    fd.write(js)
    fd.close()

elif choice == 2:
    id = input("\nEnter Product ID: ")
    print("\n       Existing Data")
    print("-------------------------------")
    print("Product Name:       ", my_record[id]['product'])
    print("Product Size:       ", my_record[id]['size'])
    print("Product For:        ", my_record[id]['gender'])
    print("Product Price:      ", my_record[id]['price'])
    print("Remaining Products: ", my_record[id]['stock'])
    print("-------------------------------\n")

    print("Choose Among: \n 1. Update Product Name \n 2. Update Product Size\n 3. Update Product for Gender\n 4. Update Product Price\n 5. Update Product Stock Number\n")
    choice = int(input("Enter your choice: "))

    if choice == 1:
        my_record[id]['product'] =input('\nEnter updated Product Name: ')

    elif choice == 2:
        my_record[id]['size'] =input('Enter updated Product Size: ')
    
    elif choice == 3:
        my_record[id]['gender'] =input('Enter updated Product for Gender: ')
    
    elif choice == 4:
        my_record[id]['price'] =input('Enter updated Product Price: ')
    
    elif choice == 5:
        my_record[id]['stock'] =input('Enter updated Product Stock Number: ')
    
    else:
        print("Invalid Choice")
    
    js = json.dumps(my_record)

    fd = open("records.json", "w")
    fd.write(js)
    fd.close()

else:
    print("Invalid Choice")
```

```python
# Purchase products

import time

id = input("Enter Product ID: ")
quantity = input("Enter Product Quantity: ")

if int(my_record[id]['stock']) == 0:
    print("\nStock Unavailable\n")

else:
    remaining_stock = int(my_record[id]['stock']) - int(quantity)
    total_cost = int(quantity) * int(my_record[id]['price'])

    print("\n*******************************")
    print("    Welcome to Fashionistas    ")
    print("*******************************")
    print("       You Have Purchased      ")
    print("-------------------------------")
    print("Product Name:       ", my_record[id]['product'])
    print("Product Size:       ", my_record[id]['size'])
    print("Product For:        ", my_record[id]['gender'])
    print("Product Price:      ", my_record[id]['price'])
    print("_______________________________")
    print("Total Cost: ", total_cost)
    print("-------------------------------\n")

    my_record[id]['stock'] = str(remaining_stock)

    js = json.dumps(my_record)

    fd = open("records.json", "w")
    fd.write(js)
    fd.close()

    fd = open("sales.json", "r")
    text = fd.read()
    fd.close()

    sales_record = json.loads(text)

    last_transaction_id = list(sales_record.keys())[-1]

    sales_record[int(last_transaction_id)+1] = {
        'product_id': id,
        'cost_per_product': my_record[id]['price'],
        'items_sold': quantity,
        'total_cost': total_cost,
        'remaining_stock': remaining_stock,
        'purchase_time' : time.ctime()
    }

    js1 = json.dumps(sales_record)

    fd = open("sales.json", "w")
    fd.write(js1)
    fd.close()
```
